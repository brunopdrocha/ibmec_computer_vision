```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carregar a imagem
img = cv2.imread('bola.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Aplicar filtro Gaussiano para suavizar a imagem
img_blur = cv2.GaussianBlur(img, (5, 5), 0)
```

```python
# Aplicar filtro mediano
img_median = cv2.medianBlur(img, 5)
```

```python
# Sobel em X (detecção de bordas horizontais)
sobel_x = cv2.Sobel(img_blur, cv2.CV_64F, 1, 0, ksize=3)

# Sobel em Y (detecção de bordas verticais)
sobel_y = cv2.Sobel(img_blur, cv2.CV_64F, 0, 1, ksize=3)

# Calculando a magnitude do gradiente (magnitude das bordas)
sobel_edges = cv2.convertScaleAbs(cv2.magnitude(sobel_x, sobel_y))
```

```python
# Aplicando Canny
canny_edges = cv2.Canny(img_blur, 30, 100)

_, canny_edges = cv2.threshold(canny_edges, 50, 255, cv2.THRESH_BINARY)
```

```python
# Displaying the results
plt.figure(figsize=(10, 8))

# Original image
plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

# Image smoothed with Gaussian filter
plt.subplot(2, 2, 2)
plt.imshow(img_blur, cmap='gray')
plt.title('Gaussian Filter')

# Edges detected with Sobel
plt.subplot(2, 2, 3)
plt.imshow(sobel_edges, cmap='gray')
plt.title('Edges with Sobel')

# Edges detected with Canny
plt.subplot(2, 2, 4)
plt.imshow(canny_edges, cmap='gray')
plt.title('Edges with Canny')

plt.tight_layout()
plt.show()
```
